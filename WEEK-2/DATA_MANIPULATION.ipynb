```python
#Data Manipulation
import pandas as pd
data={
    "name":["A","B","C"],
    "age":[20,212,22],
    "marks":[70,80,90],
}
df=pd.DataFrame(data)
print(df)
df["Bonus"]=5
print("\n after adding Bonus:")
print(df)
df["marks"]=df["marks"]+2
print("\n after increasing marks:")
print(df)
df.loc[0,"age"]=21
print("\n after changing age of first student:")
print(df)
df.rename(columns={"marks":"final"},inplace=True)
print("\n after renaming the column:")
print(df)
df.drop("Bonus",axis=1,inplace=True)
print("\n after removing Bonus:")
print(df)
```
