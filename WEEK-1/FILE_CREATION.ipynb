```python
 #Create and write to a file
file_path = "sample_text.txt"

with open(file_path, "w") as file:
    file.write("Hello World\n")
    file.write("This is a Python file handling example.\n")
    file.write("We are counting lines, words, and characters.\n")
print("File created and data written successfully.\n")
# Read the file
with open(file_path, "r") as file:
    content = file.read()
print("File Content:\n")
print(content)
lines = content.split("\n")
num_lines = len([line for line in lines if line.strip() != ""])
num_words = len(content.split())
num_characters = len(content)
print("Number of lines:", num_lines)
print("Number of words:", num_words)
print("Number of characters:", num_characters)
```
