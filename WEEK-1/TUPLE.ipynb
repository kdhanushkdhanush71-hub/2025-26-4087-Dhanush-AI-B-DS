```python
 #TUPLE OPERATIONS

# Creating tuples
t = (10, 20, 30, 40)
print("Original tuple:", t)

# Accessing elements
print("First element:", t[0])
print("Last element:", t[-1])

# Slicing
print("Sliced tuple:", t[1:4])

# Length
print("Length of tuple:", len(t))

# Membership
print("Is 30 in tuple?", 30 in t)

# Counting elements
print("Count of 20:", t.count(20))

# Finding index
print("Index of 40:", t.index(40))

# Looping
print("Tuple elements:")
for item in t:
    print(item)

# Concatenation
t2 = (50, 60)
t3 = t + t2
print("After concatenation:", t3)

# Repetition
t4 = t * 2
print("After repetition:", t4)

# Nested tuple
nested = (1, (2, 3), (4, 5))
print("Nested tuple element:", nested[1][0])

# Converting tuple to list (to modify)
lst = list(t)
lst.append(100)
t_new = tuple(lst)
print("Modified tuple (via list):", t_new)
```
