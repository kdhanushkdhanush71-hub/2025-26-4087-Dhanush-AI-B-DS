```python
import pandas as pd
import numpy as np

# Example DataFrame
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, 7, 8, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Replace missing values with column mean
df_filled = df.fillna(df.mean())

print("\nDataFrame after replacing NaN with column mean:")
print(df_filled)
```
